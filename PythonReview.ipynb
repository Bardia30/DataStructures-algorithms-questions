```python
#create array , 1000 numbers chosen randomly from normal dist, mean 5, std 2

normalArray = np.random.normal(5, 2, 1000)
print(normalArray.std())

stdArray = normalArray.std()
meanArray = normalArray.mean()

count = 0 

for num in normalArray:
    if num > 2*stdArray + meanArray:
        count += 1
print(count)
```

```python
#creating cluster points example 3.7.1

n = 200 

x1  = np.random.uniform(low=0, high=5, size=n)
y1 = np.random.uniform(low=0, high=5, size=n)

x2 = np.random.uniform(low=10, high=15, size=n)
y2 = np.random.uniform(low=10, high=15, size=n)


plt.scatter(x1, y1, color='blue', label="first cluster in [0, 5] x [0, 5]")
plt.scatter(x2, y2, color='red', label = 'second cluster in [10, 15] x [10, 15]')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.axis('equal')

plt.show()
```

```python
#circular cluster points 

n = 200 

theta1 = np.random.uniform(0, 2*np.pi, n)
theta2 = np.random.uniform(0, 2*np.pi, n)

r1 = np.random.uniform(0, 2, n)
r2 = np.random.uniform(0, 2, n)

x1 = 10 + r1 * np.cos(theta1)
y1 = 10 + r1 * np.sin(theta1)
x2 = 20 + r2 * np.cos(theta2)
y2 = 20 + r2 * np.sin(theta2)

plt.scatter(x1, y1, color='blue', label="first cluster")
plt.scatter(x2, y2, color='red', label = 'second cluster')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.axis('equal')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Number of points in each cluster
n_points = 200

# Generate random angles
theta1 = np.random.uniform(0, 2*np.pi, n_points)
theta2 = np.random.uniform(0, 2*np.pi, n_points)

# Generate radii using absolute values of normally distributed numbers
# Mean = 0, Standard Deviation = 1
r1 = np.abs(np.random.normal(0, 1, n_points))
r2 = np.abs(np.random.normal(0, 1, n_points))

# Convert polar coordinates to Cartesian coordinates and shift centers
cluster1_x = 5 + r1 * np.cos(theta1)
cluster1_y = 5 + r1 * np.sin(theta1)
cluster2_x = 10 + r2 * np.cos(theta2)
cluster2_y = 10 + r2 * np.sin(theta2)

# Create the plot
plt.figure(figsize=(8, 8))
plt.scatter(cluster1_x, cluster1_y, color='blue', label='Cluster Centered at (5,5)')
plt.scatter(cluster2_x, cluster2_y, color='red', label='Cluster Centered at (10,10)')
plt.title('Two Clusters with Normally Distributed Radii')
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.legend()
plt.grid(True)
plt.axis('equal')  # Ensures that scaling is equal on both axes

# Show the plot
plt.show()
```
